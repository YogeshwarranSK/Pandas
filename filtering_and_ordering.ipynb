FILTERING AND ORDERING

```python
import pandas as pd
```

```python
df = pd.read_csv(r"D:\Notes\Numpy & Pandas\Pandas\world_population.csv")
df
```

```python
# Using comparison operators
df[df['Rank'] <= 10]
```

```python
# Using .isin() function
specific_countries = ['Brazil','India']
df[df['Country'].isin(specific_countries)]
```

```python
# using contains
df[df['Country'].str.contains('United')]
```

```python
df2 = df.set_index('Country')
df2
```

```python
df2.filter(items=['Capital','CCA3'],axis = 1)
```

```python
df2.filter(items=['Zimbabwe'],axis=0)
```

```python
df2.filter(like = 'United',axis =0)
```

```python
df2.loc['United States']
```

```python
df2.iloc[3]
```

ORDER BY 

```python
df[df['Rank'] > 10].sort_values(by='Rank',ascending=False)
```

```python
df[df['Rank'] > 10].sort_values(by=['Rank','Country'],ascending=False)
```

```python
df[df['Rank'] > 10].sort_values(by=['Rank','Country'],ascending=[True,False])
```
